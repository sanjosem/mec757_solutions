```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.gridspec as gridspec
```

```python
plt.rcParams['figure.dpi'] = 100
plt.rcParams['font.size'] = 12
```

## Exercice 2

```python
x = np.linspace(-5,5,200)
y = np.linspace(-3,3,100)
X,Y = np.meshgrid(x,y)
U = X**2 
V = X * Y**2
```

```python
plt.streamplot(X,Y,U,V)
plt.title('Exercice 2.a')
plt.xlabel('x')
plt.ylabel('y')
plt.axis('equal')
plt.savefig('ex_2a.png')
plt.show()
```

```python
x = np.linspace(-5,5,200)
y = np.linspace(-3,3,100)
X,Y = np.meshgrid(x,y)
U = -X
V = Y
```

```python
plt.streamplot(X,Y,U,V)
plt.title('Exercice 2.b')
plt.xlabel('x')
plt.ylabel('y')
plt.axis('equal')
plt.savefig('ex_2b.png')
plt.show()
```

```python
x = np.linspace(-5,5,200)
y = np.linspace(-3,3,100)
X,Y = np.meshgrid(x,y)
R = np.sqrt(X**2+Y**2)
THETA = np.arctan2(Y,X)
Vtheta = 1./(2*np.pi*R)
U = - Vtheta * np.sin(THETA)
V = Vtheta * np.cos(THETA)
```

```python
plt.streamplot(X,Y,U,V)
plt.title('Exercice 2.c')
plt.xlabel('x')
plt.ylabel('y')
plt.axis('equal')
plt.savefig('ex_2c.png')
plt.show()
```
